## Valdidating and Cleaning Data

    Checking the uniqness of the "personID" filed as it is UUID.

    Checking the uniqness of the "email" field as it is uniq per user.

```python
def ensure_unique_personId(persons_df):
    id_map = {}  # Dictionary to keep track of existing and new IDs
    unique_ids = []
    for person_id in persons_df['personId']:
        
        if pd.isna(person_id) or id_map.get(person_id,0) != 0:
            print("found a NAN or not Uniq personID")
            new_id = str(uuid.uuid4())
            while id_map.get(new_id,0) != 0:
                new_id = str(uuid.uuid4())
            unique_ids.append(new_id)
            print("new ID",new_id)
            id_map[new_id] = True
        else:
            unique_ids.append(person_id)
            id_map[person_id] = True
    
    return unique_ids
```

```python
persons_data = persons_data.copy()
persons_data.loc[:, 'personId'] = ensure_unique_personId(persons_data)
```

## Merge occupancy_data with inventory_data to get bedId

## Merge the above result with persons_data to get all required person information

## Extractction and formating of major IDs
    Create a hashmap to map major names to the majorID 
    Perform a row operation to map list of majors to the majorId using hashmap

```python
majors_Hmap = pd.Series(majors_data.displayId.values, index=majors_data.name).to_dict()
```

```python
majors_Hmap
```

```python
def get_major_ids(majors):
    if pd.isna(majors) or majors is None:
        return " "
    else:
        major_names = str(majors).split(',')
        mjor_id_ls = [majors_Hmap.get(i.strip(), '') for i in major_names]
        return ','.join(map(str,mjor_id_ls))


persons_occupancy_inventory_merged['majorIds'] = persons_occupancy_inventory_merged['majors'].apply(lambda x: get_major_ids(x))
```

```python
persons_occupancy_inventory_merged.head()
```

## Creating address1, address2, city, state, zip'
    As we previouly mapped the addres partition we do not need a dynamic code to poppulate these columns as All addres just contains Block number & Street address, City and State.
    
    Poppulating addres1, city and state columns from address column from persons_data.

```python
def split_address(address):
    
    if pd.isna(address) or address is None:
        return pd.Series([None, None, None, None, None])
        
    parts = address.split(',')
    if len(parts) >= 3:
        address1 = parts[0].strip()
        city = parts[1].strip()
        state = parts[2].strip()
        return pd.Series([address1, None, city, state, None])
    else:
        return pd.Series([None, None, None, None, None])

```

```python
address_parts = persons_occupancy_inventory_merged['address'].apply(split_address)
address_parts.columns = ['address1', 'address2', 'city', 'state', 'zip']
```

```python
address_parts
```

## Merging the splitted address back to our main dataframe containing all the columns

```python
persons_occupancy_inventory_merged = pd.concat([persons_occupancy_inventory_merged, address_parts], axis=1)
```
